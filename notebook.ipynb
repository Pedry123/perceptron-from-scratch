```python
import numpy as np
import pandas as pd
import random
```

### Pré-processando os dados

```python
train_df = pd.read_csv('train.csv')
x = train_df[['x1', 'x2', 'x3']].to_numpy()
d = train_df['d'].to_numpy()
d = d.reshape(d.shape[0], 1).transpose()
x = x.transpose()
x = np.insert(x, 0, -1 * np.ones(x.shape[1]), axis=0)
x.shape
```

### Vetor Peso

```python
w = np.ones(x.shape[0]) * [random.random() for _ in range(x.shape[0])]
w = np.reshape(w, (w.shape[0], 1))
w.shape
```

### Hiperparâmetros

```python
args = {
    'lr': 0.1
}
```

```python
w
```

### Treinamento

```python
num_epochs = 0
weight_results = []
while True:
    error = False
    for i in range(d.shape[1]):
        u = np.dot(w.T, x[:, i]) # seleciona-se a coluna, representativa de cada amostra
        y_pred = np.sign(u)

        if y_pred[0] != d[:, i][0]:
            w += args['lr'] * (d[:,i] - y_pred[0]) * x[:, i].reshape(-1, 1) # para que o x seja vetor coluna também, para a soma funcionar
            error = True
    num_epochs += 1
    weight_results.append(w.copy())
    if not error:
        break
print(num_epochs)
print(weight_results)
```

### Fase de Operação

```python
for i in range(d.shape[1]):
    u = w.T @ x # operador matricial do numpy além de np.dot
    y_pred = np.sign(u)
    print(f"Amostra {i+1} pertencente à classe {'C1' if y_pred[0, i] < 0 else 'C2'}.")

```

```python
u = w.T @ x
y_pred = np.sign(u).ravel()

c1 = list(filter(lambda x: x[1] < 0, enumerate(y_pred)))
c2 = list(filter(lambda x: x[1] > 0, enumerate(y_pred)))
print("Amostras pertencentes à classe C1:", *list(map(lambda x: x[0]+1, c1)))
print(f"Amostras pertencentes à classe C2:", *list(map(lambda x: x[0]+1, c2)))
```
